If you don't use `test_` in the beginning of the name of your functions, it will not be counted as a test method

--- 

You can use additional tests inside of the same testing method to make sure it accepts different cases
```
def test_add(self):
    self.assertEqual(calc.add(10,5), 15)
    self.assertEqual(calc.add(-1,1), 0)
    self.assertEqual(calc.add(-1,-1), -2)

```

--- 

It is a good practice to add new tests for every bug or problem that you had, so you can assure it will not happen again

---

`self.assertRaises(ValueError, calc.divide, 10, 0)`

The method is passed as a parameter because if you use as `calc.divide(10,0)` it will rise an error and the unit test will fail

---

You can also use the context manager and call it the normal way

```
# Calling the funcion normally
with self.assertRaises(ValueError):
    calc.divide(10,0)
```

*Set ups and tear downs:*

```
def setUp(self):
    pass

def tearDown(self):
    pass


@classmethod
def setUpClass(cls):
    print('Set Up Class')

@classmethod
def tearDownClass(cls):
    print('Tear Down Class')

```

- `setUp` executes **every time** before a test starts.
- `tearDown` executes **every time** a test ends.
- `setUpClass` **runs once** before all tests.
- `tearDownClas` **runs once** after all tests.
